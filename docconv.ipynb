<div dir='rtl'>
    <h2 style='color:hotpink;'>الخطوة 1: استيراد المكتبات</h3>
    <p>نستورد <code>mistralai</code> للتفاعل مع الـ API، و<code>base64</code> لترميز الملفات قبل إرسالها عبر الشبكة.</p>
</div>

```python
import base64
from mistralai import Mistral
```

<div dir='rtl'>
    <h2 style='color:hotpink;'>الخطوة 2: قراءة الملف بصيغة ثنائية</h3>
    <p>نفتح ملف الـ PDF ونقرأه كبايتات (bytes) باستخدام الوضع <code>'rb'</code> (read binary). هذا الإجراء ضروري للتعامل مع الملفات غير النصية مثل PDF.</p>
</div>

```python
# تأكد من وجود ملف باسم "document.pdf" في نفس المجلد
with open("document.pdf", "rb") as f:
    pdf_bytes = f.read()
```

<div dir='rtl'>
    <h2 style='color:hotpink;'>الخطوة 3: ترميز البيانات (Base64)</h3>
    <p>نحوّل بايتات الملف إلى نص بترميز Base64. هذه الصيغة آمنة ومناسبة لنقل البيانات الثنائية داخل طلبات الـ API النصية.</p>
</div>

```python
b64 = base64.b64encode(pdf_bytes).decode('utf-8')
```

<div dir='rtl'>
    <h2 style='color:hotpink;'>الخطوة 4: تهيئة عميل الـ API</h3>
    <p>نُنشئ كائنًا من <code>Mistral</code> ونمرر له مفتاح الـ API الخاص بنا للمصادقة على الطلبات.</p>
</div>

```python
client = Mistral(api_key="97ZQlsV45YrDusgZRwjArWGbh3nerFPb")
```

<div dir='rtl'>
    <h2 style='color:hotpink;'>الخطوة 5: إرسال طلب الـ OCR</h3>
    <p>نستدعي دالة <code>client.ocr.process</code> ونحدد الموديل <code>mistral-ocr-latest</code>. يتم إرسال الملف المرمّز كـ <code>Data URL</code>، وهو تنسيق قياسي لتضمين البيانات في رابط.</p>
</div>

```python
resp = client.ocr.process(
    model="mistral-ocr-latest",
    document={
        "type": "document_url",
        "document_url": f"data:application/pdf;base64,{b64}"
    },
    include_image_base64=True
)
```

<div dir='rtl'>
    <h2 style='color:hotpink;'>الخطوة 6: عرض النتائج</h3>
    <p>الاستجابة (<code>resp</code>) تحتوي على قائمة بالصفحات (<code>pages</code>). نمرّ عليها في حلقة تكرارية ونطبع محتوى كل صفحة بصيغة Markdown، والتي تحافظ على التنسيق الأساسي للنص.</p>
</div>

```python
for page in resp.pages:
    print(page.markdown)
```
